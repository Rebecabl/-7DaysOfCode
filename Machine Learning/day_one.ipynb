# **Verificando valores nulos**

```python
data.isnull().sum()
```

# **Preenchendo esses valores ausentes com o valor mais frequente (moda) em cada coluna**

```python
data['artists'].fillna(data['artists'].mode()[0], inplace=True)
data['album_name'].fillna(data['album_name'].mode()[0], inplace=True)
data['track_name'].fillna(data['track_name'].mode()[0], inplace=True)
```

# **Verificando novamente se ainda há valores ausentes para garantir que todos tenham sido tratados**

```python
data.isnull().sum()
```

Não há nenhum valor nulo mais

# **Quantidade de duplicatas**

linhas ou entradas que têm exatamente os mesmos valores em todas as colunas, ou em uma combinação específica de colunas.









```python
duplicatas = data.duplicated().sum()
print("\nNúmero de Duplicatas:", duplicatas)
```

Não há duplicatas

OBS: não há duplicatas gerais, porém pode haver duplicatas em outras colunas, já que a análise é do Spotify, e um artista pode ter várias músicas. Além disso, pode haver músicas com os mesmos nomes, por exemplo, então não é algo que atrapalhe a análise.


```python
# Iterar sobre todas as colunas do DataFrame
for coluna in data.columns:
    # Verificar se há duplicatas na coluna
    duplicatas_na_coluna = data[data.duplicated(subset=[coluna], keep=False)]

    # Imprimir resultados para a coluna atual
    print(f"Coluna: {coluna}")
    print(f"Total de Duplicatas: {duplicatas_na_coluna.shape[0]}\n")
```

# **Filtrar as Músicas Mais Populares**

* Isso indica que, em média, as músicas não explícitas tendem a ter uma duração ligeiramente maior do que as músicas explícitas.
